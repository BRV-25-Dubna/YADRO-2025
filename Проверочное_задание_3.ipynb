### 2. Вычисление собственных значений и векторов

$$
T \cdot v = \lambda \cdot v
$$

где:
- $ \lambda $ — собственное значение,
- $ v $ — собственный вектор.

Это эквивалентно решению уравнения:

$$
(T - \lambda I) \cdot v = 0
$$

где $ I $ — единичная матрица.

```python
# Матрица переходов
T = np.array([[0.99, 0.2],
              [0.01, 0.8]])

# Вычисление собственных значений и векторов
eigenvalues, eigenvectors = np.linalg.eig(T)

print("Собственные значения:")
print(eigenvalues)

print("\nСобственные векторы:")
print(eigenvectors)
```

```python
# Нормализация собственных векторов
normalized_eigenvectors = []
for vector in eigenvectors.T:
    norm = np.sum(vector)
    normalized_vector = vector / norm
    normalized_eigenvectors.append(normalized_vector)

normalized_eigenvectors = np.array(normalized_eigenvectors).T

print("\nНормализованные собственные векторы:")
print(normalized_eigenvectors)
```

---

#### Собственное значение $ \lambda_1 = 1 $:

- Это означает, что матрица $ T $ имеет **стационарное состояние**, которое соответствует собственному вектору $ v_1 $.
- Собственный вектор $ v_1 $ — это **стационарное распределение исследователей**. Его можно нормализовать, чтобы получить:

$$
v_1 = \begin{bmatrix}

0.04761905
\end{bmatrix}
$$

- Это означает, что независимо от начального распределения, система сходится к стационарному состоянию, где:
  - 95.24% исследователей работают в промышленности,
  - 4.76% исследователей работают в академических кругах.

---

#### Собственное значение $ \lambda_2 = 0.79 $:

- Это второе собственное значение, которое **меньше 1**.
- Оно описывает **переходный режим**, который затухает со временем (поскольку $ 0.79^n \to 0 $ при $ n \to \infty $).
- Переходный режим постепенно исчезает, и система стабилизируется в стационарном состоянии.

### Сжатие изображения с использованием SVD

Мы уже видели, что SVD позволяет разложить любую матрицу $ A $ размером $ m \times n $ на произведение ортогональных матриц $ U $ и $ V $, а также "диагональной" матрицы $ \Sigma $:

$$
A_{m \times n} = U_{m \times m} \Sigma_{m \times n} V_{n \times n}^T
$$

Мы также обсуждали, что можно аппроксимировать исходную матрицу $ A $, используя только часть матриц $ U $, $ V $ и $ \Sigma $:

$$
A_{m \times n} \approx U_{m \times k} \Sigma_{p \times k} V_{n \times k}^T
$$

Вы можете заметить, что эта аппроксимация может помочь нам значительно сэкономить место. Действительно, вместо хранения $ m \times n $ значений в исходной матрице $ A $, теперь нам нужно хранить только $ m \times k + k + k \times n = k \times (m + n + 1) $ значений. **Поскольку изображения могут быть представлены в виде матриц пиксельных значений, мы можем использовать эту идею для сжатия изображений с помощью SVD!**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import urllib.request
```

Сначала загрузим изображение в массив $\texttt{numpy}$ и отобразим его.

Вы можете использовать *любое* изображение на ваш выбор или загрузить одно из образцовых изображений, предоставленных [в нашем репозитории](https://github.com/girafe-ai/math-basics-for-ai/tree/master/graded%20assignments/img).

```python
path = "https://drive.google.com/file/d/1P1E0mvd1eJ2t_DfQLv0LcnoJRQjpgCcB/view?usp=drive_link"
url = "https://drive.google.com/uc?id=" + path.split("/")[-2]

with urllib.request.urlopen(url) as url_response:
    img_rgb = Image.open(url_response)

img_rgb = np.array(img_rgb)

plt.figure(figsize=(5, 10))
plt.axis('off')
plt.imshow(img_rgb)
```

Обратите внимание, как изображение представлено в формате RGB:

```python
print(img_rgb.shape)
img_rgb
```

Для простоты преобразуем наше изображение в черно-белый формат (таким образом, нам нужно будет работать только с одним каналом). Мы сделаем это, усредняя сигналы из красного, зеленого и синего каналов для каждого пикселя.

```python
img_grayscale = np.mean(img_rgb, axis=-1)
```

Теперь наше изображение представлено в виде матрицы размером $ m \times n $:

```python
print(img_grayscale.shape)
img_grayscale
```

***

Мы можем построить черно-белое изображение:

```python
plt.figure(figsize=(5, 10))
plt.axis('off')
plt.imshow(img_grayscale, cmap='gray')
```

Примените SVD к матрице изображения $ M $, разложив её на произведение ортогональных матриц $ U $ и $ V $, а также "диагональной" матрицы $ \Sigma $:

$$
M_{m \times n} = U_{m \times m} \Sigma_{m \times n} V_{n \times n}^T
$$

*Вы можете использовать [$\texttt{np.linalg.svd()}$](https://numpy.org/doc/stable/reference/generated/numpy.linalg.svd.html) для выполнения SVD.*

```python
U, sigma, V = np.linalg.svd(img_grayscale, full_matrices=False)
```

***

**(2 балла)** Продемонстрируйте, как можно восстановить матрицу изображения из произведения $ U $, $ V $ и $ \Sigma $ *(выразите матрицу изображения $ M $ как произведение этих трёх матриц и постройте результирующее изображение, убедившись, что оно совпадает с исходным)*.

```python
j = 0
for r in (5, 20, 100):
  sigma_r = np.diag(sigma[:r])
  Xapprox = U[:, :r] @ sigma_r @ V[:r, :]
  plt.figure(j+1)
  j += 1
  img = plt.imshow(Xapprox)
  img.set_cmap('gray')
  plt.axis('off')
  plt.title('r = ' + str(r))
  plt.show()
```

```python
sigma_matrix = np.diag(sigma)
img_reconstructed = U @ sigma_matrix @ V

plt.figure(figsize=(5, 10))
plt.axis('off')  # Убираем оси координат
plt.imshow(img_reconstructed, cmap='gray')
```

```python
print(f"Ранг матрицы: {len(sigma)}")
```

Теперь давайте сжимать исходное изображение! Помните, что с помощью SVD можно аппроксимировать исходную матрицу, учитывая только $ p $ скрытых признаков:

$$
M_{m \times n} \approx U_{m \times p} \Sigma_{p \times p} V_{n \times p}^T
$$

**(3 балла)** Попробуйте восстановить изображение, используя только $ 5, 10, ..., 100 $ скрытых признаков. Постройте, что происходит.

*Подсказка: используйте [$\texttt{plt.subplot()}$](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html), чтобы разместить несколько изображений на одном графике.*

```python
p_values = list(range(5, 101, 5))

plt.figure(figsize=(15, 20))

for i, p in enumerate(p_values):
    # Восстановление изображения с использованием первых p сингулярных значений
    sigma_p = np.diag(sigma[:p])
    img_compressed = U[:, :p] @ sigma_p @ V[:p, :]

    plt.subplot(5, 4, i + 1)
    plt.imshow(img_compressed, cmap='gray')
    plt.title(f'p = {p}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

Уже при $ p = 40 $ изображение становится достаточно чётким, а дальнейшее восстановление вносит незначительные изменения.
