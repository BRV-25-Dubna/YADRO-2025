# Итоговый проект.

Анализ личности клиента — это подробный анализ клиентов компании. Он помогает компании лучше понять своих клиентов и облегчает изменение продукции в соответствии со специфическими потребностями, поведением различных типов клиентов.

Анализ личности клиента помогает компании модифицировать свой продукт в соответствии с целевыми потребителями из различных сегментов потребителей. Например, вместо того, чтобы тратить деньги на маркетинг нового продукта для каждого клиента в базе данных компании, компания может проанализировать, какой сегмент клиентов с наибольшей вероятностью купит продукт, и затем продвигать продукт только для этого сегмента.

Описание столбцов:

```python
# Определение границ IQR для каждого числового столбца
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Определение границ для определения выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ['Income', 'Year_Birth', 'NumCatalogPurchases', 'NumWebPurchases']

# Удаление выбросов из данных
for outlier in outliers:
    df = df[(df[outlier] >= lower_bound[outlier]) 
            & (df[outlier] <= upper_bound[outlier])]
    
# Удаление максимально значения по столбцу MntMeatProducts
df = df.drop(df['MntMeatProducts'].idxmax())

# Переиндексация датафрейма
df = df.reset_index(drop=True)
```

##### **Вопрос 2:**
Какой средний возраст покупателя? Зависит ли возраст клиента от его покупок во всех категориях?
- Средний возраст клиента `~50` лет.
- Возраст никак не коррелирует с покупками.

```python
# Создание полного мультииндекса всех комбинаций
multi_index = pd.MultiIndex.from_product([df['Kidhome'].unique(), 
                                          df['Teenhome'].unique()], 
                                         names=['Kidhome', 'Teenhome'])

# Группировка данных по наличию детей и подростков
grouped = df.groupby(['Kidhome', 'Teenhome'])[
    ['MntWines', 'MntFruits', 'MntMeatProducts', 
     'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()

# Переиндексация группированного DataFrame
grouped = grouped.reindex(multi_index, fill_value=0)

# Создание нового DataFrame для таблицы
table = pd.DataFrame({
    'Нет детей': 
        grouped[(grouped.index.get_level_values('Kidhome') + 
                 grouped.index.get_level_values('Teenhome')) == 0].sum(),
    'Один ребёнок': 
        grouped[(grouped.index.get_level_values('Kidhome') + 
                 grouped.index.get_level_values('Teenhome')) == 1].sum(),
    'Два ребёнка': 
        grouped[(grouped.index.get_level_values('Kidhome') + 
                 grouped.index.get_level_values('Teenhome')) > 1].sum()})

# Вывод таблицы
display(table)
```
